```python
#From a file contaning numbers seprated by commas, print the count of even numbers

count = 0
with open("demo.txt", "r") as f:
    data = f.read()
    nums = data.split(",")
    for val in nums:
        if(int(val)%2 == 0):
            count +=1

print(count)
```
