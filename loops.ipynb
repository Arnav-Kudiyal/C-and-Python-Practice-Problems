```python
# Print the numbers from 1 to 100

i = 1
while(i<=100):
    print(i)
    i+=1
```

```python
# Print the number from 100 to 1
i = 100
while(i >=1):
    print(i)
    i-=1
```

```python

#Print the multiplication table of a numbe n

number = int(input("Please enter the number whose table you want to know"))
i = 0
while(i<=10):
    print(f"{number} * {i} = {number * i}")
    i +=1
```

```python
#Print the element of the follwoing list usingn a loop
# [1,4,9,16,25,36,49,64,81,100]

number = 1
last = 0
while(number <=10):
    
    ls = (2*number -1) + last
    print(ls) 
    number+=1
    last = ls
```

```python
# Search for a number x in the tuple using loop
#[1,4,9,16,25,36,49,64,81,100]

tup = (1,4,9,16,25,36,49,64,81,100)
number = int(input("Enter the number you want to search from the given list"))
i = 0
length = len(tup)
while(i<=length):
    if(tup[i]== number):
        print(f"The number is found at the index {i}")
        break
    else:
        i +=1
```

```python
# Print the element of the following list using a loop
#[1,4,9,16,25,36,49,64,81,100]

ls = [1,4,9,16,25,36,49,64,81,100]

for val in ls:
    print(val)
```

```python
#Search a number x in this tuple using for loop
#[1,4,9,16,25,36,49,64,81,100]

tup = (1,4,9,16,25,36,49,64,81,100)
number = int(input("enter the number you want to search"))

for val in tup:
    if(val == number):
        print(f"The number is found at index {val}")
        break
else:
    print("The number is not in the tupule")
```

#### using for & range()

```python
# Print numbers from 1 to 100
seq = range(1,101)

for val in seq:
    print(val)
```

```python
# Print numbers from 100 to 1

seq = range(100, 0, -1)

for val in seq:
    print(val)
```

```python
# Print the miltiplication table of a number n

seq = range(0,11)
i = int(input("Please enter the number whose table you want to get"))
for val in seq:
    print(f"{i} * {val} = {i*val}")
 
```

```python
# WAP to find the sum of the first n numbers.(using while)
n = int(input("Please enter the range till you want to perform sum"))
sum = 0 
i = 1
while(i<=n):
    sum = sum + i
    i+=1
print(f"The sum is {sum}")
```

```python
# WAP to find the factorial of the first n numbers.(using for)
n = int(input("Please enter the range till you want to perform factorial"))
fact = 1
for i in range(1,n+1):
    fact*=i

print(fact)
```
