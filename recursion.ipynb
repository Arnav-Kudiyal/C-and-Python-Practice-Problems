```python
#Find the sum of first n natural numbers

n = 5
def rec_sum(n):
    if(n==0):
        return 0
     
    return rec_sum(n-1) + n

print(rec_sum(n)) 
```

```python
#Find the factorial of n

n = 5
def fact(n):
    if(n==1 or n==0):
        return 1
    return fact(n-1) * n
print(fact(n))
```
