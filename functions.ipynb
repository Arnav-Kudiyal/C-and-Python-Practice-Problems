```python
# Factorial using recursion

n = int(input("Please Enter the number whose factorial you want to get"))

def fact(n):
        if(n==1 or n==0):
                return 1
        else:
            return fact(n-1) * n

print(fact(5))
```

```python
# WAF a recursive function to calculate the sum of first n natural numbers

n = int(input("Please enter the range till which you want to find the sum"))

def rec_sum(n):
    if(n==0):
        return 0
    else:
        return rec_sum(n-1) + n
  
    

print(rec_sum(n))
```
